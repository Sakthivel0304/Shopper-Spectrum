```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


df = pd.read_csv("online_retail.csv")

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
# Pivot table: rows = CustomerID, columns = Description, values = Quantity purchased
product_matrix = df.pivot_table(index='CustomerID', columns='Description', values='Quantity', fill_value=0)

product_matrix.head()
```

```python
# Cosine similarity between products (columns)
product_similarity = cosine_similarity(product_matrix.T)

# Convert to DataFrame for easy lookup
similarity_df = pd.DataFrame(product_similarity, index=product_matrix.columns, columns=product_matrix.columns)

similarity_df.head()
```

```python
def get_similar_products(product_name, top_n=5):
    if product_name not in similarity_df.columns:
        return "Product not found in dataset."
    
    # Get similarity scores
    similar_scores = similarity_df[product_name]
    
    # Sort descending, exclude the product itself
    similar_scores = similar_scores.sort_values(ascending=False).drop(product_name)
    
    # Return top N similar products
    return similar_scores.head(top_n)

# Example usage
get_similar_products("WHITE HANGING HEART T-LIGHT HOLDER")
```

```python
import joblib
joblib.dump(similarity_df, "product_similarity_matrix.pkl")
```
